# show the relation about ecletricity tax for each region

```python
df_el_re=df_2016.loc[:,['Nom région','Taux de chauffage électrique']]
df_el_re=df_el_re.groupby('Nom région').mean()
#df_el_re=sorted(df_el_re,key=df_el_re["Taux de chauffage électrique"])
df_el_re.sort_values(by=['Taux de chauffage électrique'],ascending=False)
```

```python
from matplotlib import pyplot as plt 
plt.figure(figsize=(60, 30))  
plt.xlabel('Nom région')
plt.ylabel('Taux de chauffage électrique') 
plt.bar(df_el_re.index.get_values(),df_el_re['Taux de chauffage électrique'])
plt.show()
```
